```python
# time complexity O(log(n)) space O(1)


def binary_search(word: str, words: list[str], definitions: list[str]) -> Optional[str]:
    # return the definition of the word using a binary search
    left = 0
    right = len(words)
    counter = 0
    while left < right:
        counter += 1
        win_size = right - left
        half_win = win_size // 2
        mid = left + half_win
        found_word = words[mid]
        if found_word == word:
            print(f"{counter=}")
            return definitions[mid]
        if found_word > word:
            right = mid
        else:
            left = mid + 1

    return None
```
