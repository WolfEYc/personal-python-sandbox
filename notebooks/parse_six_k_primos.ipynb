```python
import random

SEPERATOR = "|"
six_k_primos = map(lambda _: random.randint(0, 1000000), range(6000))
six_k_primos_strs = map(str, six_k_primos)
six_k_primos_piped = SEPERATOR.join(six_k_primos_strs)
six_k_primos_piped
```

```python
import time

start = time.time()
six_k_primos_parsed = SEPERATOR.split(six_k_primos_piped)
end = time.time()
duration_s = end - start
duration_ms = duration_s * 1000
duration_micros = duration_ms * 1000
duration_micros
```

```python
pct_out_of_6s = duration_s / 6 * 100
print(pct_out_of_6s, "%")
```
